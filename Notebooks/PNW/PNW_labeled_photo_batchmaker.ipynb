```python
import pandas as pd
```

```python
ob_data= pd.read_csv('../../Data/processedData/PNW/labeled_PNW_obs.csv')
```

```python
ob_data
```

```python
#We download a random sample of 50000 images
sample = ob_data.sample(50000, random_state=26)
```

```python
sample.head()
```

```python

#Here is code to create a batch file to download all the necessary images. 

n = 250
m = 200
with open("../../Data/processedData/PNW/Bats/labeledPhotos.bat", "w") as file:
    for j in range(n):
        file.write('curl')
        for i in range(m):
            idx = i + m*j 
            url = sample['photo_link'].iloc[idx]
            file_name = sample['file_name'].iloc[idx]
            path = 'D:/PNW_Angiosperms/labeled_photos/'
            full_name = path + file_name
            file.write(' -o ' + full_name + ' ' + url)
        file.write('\n')
```

```python
sample.to_csv('../../Data/processedData/PNW/PNW.csv', index=False)
```
